```python
data = dict(
        type = 'choropleth',
        locations = df_countries['Country Code'],
        z = (df_countries['Total (thousand metric tons)']),
        colorscale="Portland",
        text = df_countries['Country'],
        colorbar = {'title' : 'Title'}
      ) 
```

```python
# slider
'''
steps = []
for i in range(len(years)): 
    step = dict(method='restyle',
                args=['visible', [False] * len(years)],
                label='Year {}'.format(i + 1751))
    step['args'][1][i] = True
    i +=50
    steps.append(step)

sliders = [dict(active=0,
                pad={"t": 1},
                steps=steps)]
'''
data_slider = []
for year in df_countries['Year'].unique():
    df_year = df_countries[df_countries['Year'] == year]

    for col in df_year.columns:
        df_year[col] = df_year[col].astype(str)

    data_one_year = dict(
                        type='choropleth',
                        locations = df_year['Country'],
                        z=df_year['Total (thousand metric tons)'].astype(float),
                        locationmode='country names',
                        colorscale = "reds",
                        )

    data_slider.append(data_one_year)
    
    
steps = []
for i in range(0,len(data_slider), 20): 
    step = dict(method='restyle',
                args=['visible', [False] * len(data_slider)],
                label='Year {}'.format(i + 1751))
    #i +=50
    step['args'][1][i] = True
    #i +=50
    steps.append(step)
    #i +=50

sliders = [dict(active=0,
                pad={"t": 1},
                steps=steps)]    
```

```python
layout = dict(title = 'Fossil Fuel Emissions', 
             geo = dict(showframe=False, 
                       projection = {'type': 'mercator'}), sliders=sliders)
```

```python
choromap3 = go.Figure(data=data_slider, layout = layout)
choromap3.update_layout(
    autosize=False,
    width=800,
    height=800,)
```
